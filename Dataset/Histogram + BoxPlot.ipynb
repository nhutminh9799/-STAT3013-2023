```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Đọc dữ liệu từ file Nike.csv
df = pd.read_csv("Nike.csv")

# Lấy cột "Close"
close_values = df["Close"]

# Loại bỏ giá trị NaN từ cột "Close"
close_values = close_values.dropna()

# Tính các thống kê
count = close_values.count()
mean = close_values.mean()
variance = close_values.var()
std_dev = close_values.std()
median = close_values.median()
max_value = close_values.max()
min_value = close_values.min()
mode = close_values.mode().values[0]
quantile_1 = close_values.quantile(0.25)
quantile_3 = close_values.quantile(0.75)
coefficient_of_deviation = std_dev / mean
curtosis = kurtosis(close_values)
skewness = skew(close_values)

# Hiển thị các thống kê
print(f"Count: {count}")
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Median: {median}")
print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Mode: {mode}")
print(f"Quantile 1: {quantile_1}")
print(f"Quantile 3: {quantile_3}")
print(f"Coefficient of Deviation: {coefficient_of_deviation}")
print(f"Curtosis: {curtosis}")
print(f"Skewness: {skewness}")

# Vẽ box plot
plt.figure(figsize=(10, 6))
plt.boxplot(close_values, vert=False)
plt.title('Box Plot - Adidas Stock Price (Close)')
plt.xlabel('Close Price')
plt.show()

# Vẽ histogram
plt.figure(figsize=(10, 6))
plt.hist(close_values, bins=30, edgecolor='black')
plt.title('Histogram - Adidas Stock Price (Close)')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
# Load the dataset
df = pd.read_csv('puma.csv')
```

```python
# Select the 'Close' column and handle missing values
df = df[['Close']].dropna().reset_index(drop=True)
```

# Train-test(7-3)

```python
# Split the data into training and testing sets
train_size = int(0.7 * len(df))
test_size = int(0.3 * len(df))
train_data = df[:train_size]
test_data = df[train_size:train_size+test_size]
# Prepare training data
x_train = np.array(train_data.index).reshape(-1, 1)
y_train = np.array(train_data['Close'])
```

```python
# Train the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Prepare testing data
x_test = np.array(test_data.index).reshape(-1, 1)
y_test = np.array(test_data['Close'])
# Predict stock prices on the test set
y_pred = model.predict(x_test)

# Create an index for the next 30 days
last_index = df.index[-1]
x_next_30_days = np.array(range(last_index+1, last_index+31)).reshape(-1, 1)

# Predict the closing prices for the next 30 days
y_next_30_days = model.predict(x_next_30_days)
```

```python
# Evaluate model performance using metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(np.mean((y_pred - y_test)**2))
mape = mean_absolute_percentage_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Print metrics
print('Mean Squared Error (MSE):', mse)
print('Root Mean Squared Error (RMSE):', rmse)
print('Mean Absolute Percentage Error (MAPE):', mape)
print('Mean Absolute Error (MAE):', mae)
# Plot the predictions graph
plt.title('Linear Regression - Stock Price Prediction')
plt.plot(train_data.index, train_data['Close'], label='Train')
plt.plot(test_data.index, test_data['Close'], label='Test')
plt.plot(test_data.index, y_pred, label='Predictions')
plt.plot(range(last_index+1, last_index+31), y_next_30_days, label='Next 30 Days')
plt.legend()
plt.show()
```

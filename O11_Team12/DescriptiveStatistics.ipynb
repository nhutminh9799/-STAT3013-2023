# ACB Descriptive Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

print('-----ACB Descriptive Statistics-----')
acb_data = pd.read_csv('ACB.csv')

# Calculating count
print('Count:', len(acb_data))

# Calculating max
print('Max:', max(acb_data['Price']))

# Calculating min
print('Min:', min(acb_data['Price']))

# Calculating the mean
print('Mean:', acb_data['Price'].mean())

# Calculating the median
print('Median:', acb_data['Price'].median())

# Calculating the mode
print('Mode:', acb_data['Price'].mode())

# Calculating the standard deviation
print('Standard Deviation:', acb_data['Price'].std())

# Calculating the coeficient of deviation
print('Coefficient of Deviation:', acb_data['Price'].std()/acb_data['Price'].mean())

# Calculating the variance
print('Variance:', acb_data['Price'].var())

# Calculating skewness
print('Skewness:', acb_data['Price'].skew())

# Calculating kurtosis
print('Kurtosis:', acb_data['Price'].kurt())

# Calculating the range
print('Range:', acb_data['Price'].max() - acb_data['Price'].min())

# Calculating percentiles (0th, 25th, 50th, 75th, and 100th percentiles)
acb_quantiles = acb_data['Price'].quantile([0, 0.25, 0.5, 0.75, 1])
print('0th Percentile:', acb_quantiles[0])
print('25th Percentile (Q1):', acb_quantiles[0.25])
print('50th Percentile (Median):', acb_quantiles[0.5])
print('75th Percentile (Q3):', acb_quantiles[0.75])
print('100th Percentile:', acb_quantiles[1])

plt.figure(figsize=(10, 6))
plt.plot(acb_data['Price'], color='blue')
plt.title('ACB Stock Price Plot')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(acb_data['Price'], bins=30, color='blue')
plt.title('ACB Stock Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot(acb_data['Price'], vert=True)
plt.title('ACB Stock Price Boxplot')
plt.ylabel('Price')
plt.show()
```
